### Phase 3: Univariate Analysis

#### Cause of Accident Distribution

```python
cause_counts = project['Cause_of_accident'].value_counts().sort_values(ascending=False)  # Count accident causes
cum_pct = cause_counts.cumsum() / cause_counts.sum() * 100 # Cumulative Percentage

sns.set_style("dark")
plt.figure(figsize=(14,7))
fig, ax1 = plt.subplots(figsize=(14,7))

bars = ax1.bar(cause_counts.index, cause_counts.values, edgecolor='black', linewidth=1.2)

for bar in bars:
    height = bar.get_height()
    ax1.text(
        bar.get_x() + bar.get_width()/2,
        height + 5,
        f'{int(height)}',
        ha='center',
        fontsize=9,
        fontweight='bold'
    )

ax1.set_ylabel('Number of Accidents', fontsize=12, fontweight='bold')
ax1.set_xlabel('Cause of Accident',fontsize=12, fontweight='bold')
ax1.tick_params(axis='x', rotation=45)

ax2 = ax1.twinx()  # Cumulative Line

ax2.plot(
    cause_counts.index,
    cum_pct,
    marker='d',
    linewidth=3
)

# 80% Rule Line
ax2.axhline(
    y=80,
    linestyle='-.',
    linewidth=2,
    label='80% Threshold'
)

ax2.set_ylabel('Cumulative Percentage', fontsize=12, fontweight='bold')
ax2.set_ylim(0, 105)

for i, pct in enumerate(cum_pct):
    ax2.annotate(
        f'{pct:.1f}%',
        (i, pct),
        textcoords="offset points",
        xytext=(0,8),
        ha='center',
        fontsize=9,
        fontweight='bold'
    )

plt.title('Pareto Analysis of Road Accident Causes (80/20 Principle)', fontsize=16, fontweight='bold',pad=15)
ax2.legend(loc='center right')
plt.tight_layout()
plt.show();
```

**Insights**

- The Pareto analysis reveals that a small number of accident causes contribute to the majority of road accidents, validating the 80/20 principle in accident occurrence.
- The first few leading causes account for approximately 80% of total accidents, indicating that targeting these causes can significantly reduce overall accident rates.
- Causes appearing at the left side of the chart should be considered high-priority risk factors for road safety interventions and awareness campaigns.
- Less frequent causes contribute only marginally to the total accident burden, suggesting that resources should first focus on the dominant causes for maximum impact.
- Transportation authorities can use these findings to design targeted policies, driver training programs, and enforcement strategies aimed at the most critical accident causes, potentially achieving substantial reductions in accident frequency.

### Phase 8: Accident Floe & Relationship Analysis

#### Sankey Diagram: Cause of Accident → Accident Severity

```python
sample = (project.groupby(['Cause_of_accident', 'Accident_severity']).size().reset_index(name='Count'))
labels = list(pd.concat([sample['Cause_of_accident'],sample['Accident_severity']]).unique())
source = sample['Cause_of_accident'].apply(lambda x: labels.index(x))
target = sample['Accident_severity'].apply(lambda x: labels.index(x))

fig = go.Figure(
    go.Sankey(
        arrangement="snap",
        node=dict(
            pad=20,
            thickness=20,
            line=dict(color="black", width=0.5),
            label=labels
        ),
        link=dict(
            source=source,
            target=target,
            value=sample['Count']
        )
    )
)

fig.update_layout(title={'text': 'Flow of Accident Causes to Severity Levels','x': 0.5,'font': {'size': 22}},font_size=12,height=700)
fig.show()
```

- Thicker flows represent accident causes that generate a larger number of incidents, helping identify the most influential contributors to overall road safety risks.
- Certain causes may channel a substantial proportion of accidents toward serious or fatal injury categories, indicating high-risk factors that require immediate intervention.
- The flow structure highlights that accident severity is not evenly distributed across causes, suggesting that some accident causes are inherently more dangerous than others.

#### Network Graph: Accident Cause ↔ Severity Relationship

```python
import networkx as nx
```

```python
temp = (project.groupby(['Cause_of_accident', 'Accident_severity']).size().reset_index(name='weight'))

G = nx.Graph()

for _, row in temp.iterrows():
    G.add_edge(
        row['Cause_of_accident'],
        row['Accident_severity'],
        weight=row['weight']
    )

plt.figure(figsize=(16,12))
pos = nx.spring_layout(
    G,
    k=2,
    seed=42
)

node_sizes = [G.degree(node) * 1500 for node in G.nodes()]

edge_widths = [G[u][v]['weight'] / 50for u, v in G.edges()]

nx.draw_networkx_nodes(G,pos,node_size=node_sizes,alpha=0.9)
nx.draw_networkx_edges(G,pos,width=edge_widths,alpha=0.5)
nx.draw_networkx_labels(G,pos,font_size=9,font_weight='bold')

plt.title('Accident Cause and Severity Relationship Network',fontsize=18,fontweight='bold',pad=20)
plt.axis('off')
plt.tight_layout()
plt.show()
```

**Insights**
- The network graph visualizes the relationships between accident causes and accident severity levels, where nodes represent categories and edges represent their connections.
-  Larger nodes indicate highly connected factors, suggesting accident causes or severity levels that play a central role in the overall accident network.
- Thicker edges represent stronger relationships, highlighting accident causes that contribute substantially to specific severity outcomes.
- Central nodes within the network may act as key risk drivers, making them important targets for road safety interventions and accident prevention strategies.
- The network structure provides a holistic view of accident dynamics, enabling stakeholders to identify influential factors and prioritize resources toward the most impactful safety improvements.

### Phase 9: Machine Learning Model

#### Model Training

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

X = project.drop('Accident_severity', axis=1)
y = project['Accident_severity']

X = pd.get_dummies(X)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = RandomForestClassifier()
model.fit(X_train, y_train)
```
